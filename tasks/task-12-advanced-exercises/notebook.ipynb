## Exercise 1: HuggingFace Compatibility
Make your BERT compatible with HuggingFace transformers.

```python
from transformers import BertConfig, BertModel

config = BertConfig(
    vocab_size=30000,
    hidden_size=256,
    num_hidden_layers=4,
    num_attention_heads=4,
    intermediate_size=512
)
hf_bert = BertModel(config)
print(f"HuggingFace BERT: {sum(p.numel() for p in hf_bert.parameters()):,} parameters")
```
